```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
student1={'chinese':78,'english':80,'math':99,'history':88,'discover':99}
student2={'chinese':99,'english':88,'math':70,'history':80,'discover':78}

subjects=list(student1.keys())
value1=list(student1.values())
value2=list(student2.values())

figure=plt.figure(figsize=(8.0,5.0),dpi=250.0,facecolor="#cccccc")
ax=figure.add_subplot()
ax.plot(subjects,value1)

ax.plot(subjects,value2)
plt.show()
```

```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':90, 'english':75, 'math':98, 'history':65, 'discover':75}

subjects =list(student1.keys())
value1 =list(student1.values())
value2 = list(student2.values())

figure = plt.figure(figsize=(8.0, 5.0), dpi=250.0, facecolor="#cccccc")
ax = figure.add_subplot()
ax.plot(subjects, value1,'b-.o',linewidth=1)

ax.plot(subjects, value2,'c--^')

plt.show()
```
